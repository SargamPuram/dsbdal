```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Titanic dataset from seaborn's built-in datasets
dataset = sns.load_dataset("titanic")
```

```python
# Check the first few rows of the dataset
print(dataset.head())
```

```python

# Displot for Age
sns.displot(dataset['age'].dropna(), bins=10, kde=True)
plt.title('Displot for Age')
plt.show()
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset.isna().sum()
```

```python
# Histplot for Age without KDE
sns.histplot(dataset['age'].dropna(), bins=10, kde=False)
plt.title('Histplot for Age (No KDE)')
plt.show()
```

```python
# Jointplot (Scatter) between Age and Fare
sns.jointplot(x='age', y='fare', kind='scatter', data=dataset)
plt.title('Jointplot (Scatter)')
plt.show()
```

```python
# Jointplot (Hexagonal) between Age and Fare
sns.jointplot(x='age', y='fare', kind='hex', data=dataset)
plt.title('Jointplot (Hexagonal)')
plt.show()
```

```python
# Rugplot for Fare
sns.rugplot(dataset['fare'])
plt.title('Rugplot for Fare')
plt.show()
```

```python
# Barplot (Mean Age by Gender)
sns.barplot(x='sex', y='age', data=dataset)
plt.title('Barplot (Mean Age by Gender)')
plt.show()
```

```python
# Barplot (Standard Deviation Age by Gender)
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std)
plt.title('Barplot (Standard Deviation Age by Gender)')
plt.show()
```

```python
# Countplot (Male vs Female)
sns.countplot(x='sex', data=dataset)
plt.title('Countplot (Male vs Female)')
plt.show()
```

```python
# Boxplot (Age Distribution by Gender)
sns.boxplot(x='sex', y='age', data=dataset)
plt.title('Boxplot (Age Distribution by Gender)')
plt.show()
```

```python
# Boxplot with Hue (Survival by Gender and Age)
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
plt.title('Boxplot with Hue (Survival by Gender and Age)')
plt.show()
```

```python
# Violin Plot (Age Distribution by Gender)
sns.violinplot(x='sex', y='age', data=dataset)
plt.title('Violin Plot (Age Distribution by Gender)')
plt.show()
```

```python
# Violin Plot with Hue (Survival by Gender and Age)
sns.violinplot(x='sex', y='age', data=dataset, hue='survived')
plt.title('Violin Plot with Hue (Survival by Gender and Age)')
plt.show()
```

```python
# Stripplot (Age vs Gender without Jitter)
sns.stripplot(x='sex', y='age', data=dataset, jitter=False)
plt.title('Stripplot (Age vs Gender without Jitter)')
plt.show()
```

```python
# Stripplot (Age vs Gender with Jitter)
sns.stripplot(x='sex', y='age', data=dataset, jitter=True)
plt.title('Stripplot (Age vs Gender with Jitter)')
plt.show()
```

```python
# Stripplot with Hue (Survival by Gender and Age)
sns.stripplot(x='sex', y='age', data=dataset, jitter=True, hue='survived')
plt.title('Stripplot with Hue (Survival by Gender and Age)')
plt.show()
```

```python
# Swarmplot (Age vs Gender)
sns.swarmplot(x='sex', y='age', data=dataset)
plt.title('Swarmplot (Age vs Gender)')
plt.show()
```

```python
# Swarmplot with Hue (Survival by Gender and Age)
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
plt.title('Swarmplot with Hue (Survival by Gender and Age)')
plt.show()
```

```python
# Correlation Matrix Heatmap
numeric_cols = dataset.select_dtypes(include=[np.number])
corr = numeric_cols.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap (Correlation Matrix)')
plt.show()
```

```python
# Cluster Map (Correlation Matrix)
sns.clustermap(corr, annot=True)
plt.title('Cluster Map (Correlation Matrix)')
plt.show()
```

```python
# Histogram for Fare
sns.histplot(dataset['fare'].dropna(), kde=False, bins=10)
plt.title('Histogram for Fare')
plt.show()
```

```python
sns.histplot(data=dataset, x='fare', hue='pclass', multiple="stack")
plt.title('Fare Distribution by Pclass')
plt.show()
```

```python
sns.histplot(data=dataset, x='fare', hue='survived', multiple="stack")
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
# Plot histogram for 'fare'
sns.histplot(dataset['fare'].dropna(), kde=False, bins=10)
plt.title('Histogram of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```
